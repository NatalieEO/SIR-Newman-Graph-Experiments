```python
oneCallTrial1 = pd.read_csv('../1milData/randComparison/triFree20-1call-trial1.csv', sep=',')
oneCallTrial2 = pd.read_csv('../1milData/randComparison/triFree20-1call-trial2.csv', sep=',')

loopCallTrial1 = pd.read_csv('../1milData/randComparison/triFree20-loopTimeCalls-trial1.csv', sep=',')
loopCallTrial2 = pd.read_csv('../1milData/randComparison/triFree20-loopTimeCalls-trial2.csv', sep=',')

loopijCallTrial1 = pd.read_csv('../1milData/randComparison/triFree20-loopijCalls-trial1.csv', sep=',')
loopijCallTrial2 = pd.read_csv('../1milData/randComparison/triFree20-loopijCalls-trial2.csv', sep=',')
```

```python
oneCallTrial1['PortionInf'] = oneCallTrial1['NumInfected']/oneCallTrial1['NumNodes']
oneCallTrial1['InfRate'] =oneCallTrial1['InfRate']/10000
oneCallTrial1.tail()
```

```python
oneCallTrial2['PortionInf'] = oneCallTrial2['NumInfected']/oneCallTrial2['NumNodes']
oneCallTrial2['InfRate'] =oneCallTrial2['InfRate']/10000
oneCallTrial2.tail()
```

```python
loopCallTrial1['PortionInf'] = loopCallTrial1['NumInfected']/loopCallTrial1['NumNodes']
loopCallTrial1['InfRate'] =loopCallTrial1['InfRate']/10000

loopCallTrial2['PortionInf'] = loopCallTrial2['NumInfected']/loopCallTrial2['NumNodes']
loopCallTrial2['InfRate'] =loopCallTrial2['InfRate']/10000

loopijCallTrial1['PortionInf'] = loopijCallTrial1['NumInfected']/loopijCallTrial1['NumNodes']
loopijCallTrial1['InfRate'] =loopijCallTrial1['InfRate']/10000

loopijCallTrial2['PortionInf'] = loopijCallTrial2['NumInfected']/loopijCallTrial2['NumNodes']
loopijCallTrial2['InfRate'] =loopijCallTrial2['InfRate']/10000
```

```python
loopCallTrial1= loopCallTrial1[['InfRate', 'PortionInf']]
loopCallTrial1 = loopCallTrial1.groupby('InfRate').mean()
loopCallTrial2= loopCallTrial2[['InfRate', 'PortionInf']]
loopCallTrial2 = loopCallTrial2.groupby('InfRate').mean()

oneCallTrial1= oneCallTrial1[['InfRate', 'PortionInf']]
oneCallTrial1 = oneCallTrial1.groupby('InfRate').mean()
oneCallTrial2= oneCallTrial2[['InfRate', 'PortionInf']]
oneCallTrial2 = oneCallTrial2.groupby('InfRate').mean()

loopijCallTrial1= loopijCallTrial1[['InfRate', 'PortionInf']]
loopijCallTrial1 = loopijCallTrial1.groupby('InfRate').mean()
loopijCallTrial2= loopijCallTrial2[['InfRate', 'PortionInf']]
loopijCallTrial2 = loopijCallTrial2.groupby('InfRate').mean()
```

```python
mp.figure(figsize=(25,12))
mp.title('One srand Call with Time', fontsize=22)

mp.plot(oneCallTrial1, 'ms--', label='Trial 1')
mp.plot(oneCallTrial2, 'bs--', label='Trial 2')
#mp.plot(once, 'ks--', label='srand seeded once')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
mersenne1 = pd.read_csv('../1milData/randComparison/triFree20-Mersenne-trial1.csv', sep=',')
mersenne2 = pd.read_csv('../1milData/randComparison/triFree20-Mersenne-trial2.csv', sep=',')

mersenne1['PortionInf'] = mersenne1['NumInfected']/mersenne1['NumNodes']
mersenne1['InfRate'] =mersenne1['InfRate']/10000
mersenne2['PortionInf'] = mersenne2['NumInfected']/mersenne2['NumNodes']
mersenne2['InfRate'] =mersenne2['InfRate']/10000

mersenne1= mersenne1[['InfRate', 'PortionInf']]
mersenne1 = mersenne1.groupby('InfRate').mean()
mersenne2= mersenne2[['InfRate', 'PortionInf']]
mersenne2 = mersenne2.groupby('InfRate').mean()
```
