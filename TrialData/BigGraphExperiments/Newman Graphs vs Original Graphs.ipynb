```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
BipNewman_GS30_AD20_C007 = createChart("../1milData/giantComponent/Summer/BipNewman-1mil-GS30-AD20-C007.csv")
BipNewman_GS30_AD20_C014 = createChart("../1milData/giantComponent/Summer/BipNewman-1mil-GS30-AD20-C014.csv")
BipNewman_GS30_AD20_C021 = createChart("../1milData/giantComponent/Summer/BipNewman-1mil-GS30-AD20-C021.csv")

TriangleFree_GS30_AD20 = createChart("../1milData/giantComponent/Summer/TriFree-1mil-GS30-AD20.csv")
NoGroups_GS30_AD20 = createChart("../1milData/giantComponent/Summer/NoGroups-1mil-GS0-AD20.csv")
Triangle_GS30_AD20 = createChart("../1milData/giantComponent/Summer/Triangle-1mil-GS30-AD20-15.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Original Methods GS = 30, AD = 20 | Triangle vs Tri-Free vs No Groups ', fontsize=22)

mp.plot(TriangleFree_GS30_AD20[.1:.55], 'mo--', label='Triangle Free')
mp.plot(NoGroups_GS30_AD20[.1:.55], 'bo--', label='No Groups')
mp.plot(Triangle_GS30_AD20[.1:.55], 'go--', label='Triangle')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle-Free Newman vs Triangle-Free Original Method | GS = 30, AD = 20 ', fontsize=22)

mp.plot(BipNewman_GS30_AD20_C007, 'ms--', label='Clustering Coef = 0.07')
mp.plot(BipNewman_GS30_AD20_C014, 'b*--', label='Clustering Coef = 0.14')
mp.plot(BipNewman_GS30_AD20_C021, 'go--', label='Clustering Coef = 0.21')
mp.plot(TriangleFree_GS30_AD20, 'ko--', label='Triangle Free')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
