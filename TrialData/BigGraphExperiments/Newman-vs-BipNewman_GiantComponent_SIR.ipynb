```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;

def createChart2(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
Newman_GS10_AD5_C007 = createChart("../1milData/giantComponent/updatedGC/August/Newman-GC-1mil-GS10-AD5-C007.csv")
Newman_GS10_AD5_C014 = createChart("../1milData/giantComponent/updatedGC/August/Newman-GC-1mil-GS10-AD5-C014.csv")
Newman_GS10_AD5_C021 = createChart("../1milData/giantComponent/updatedGC/August/Newman-GC-1mil-GS10-AD5-C021.csv")
```

```python
BipNewman_GS10_AD5_C007 = createChart("../1milData/giantComponent/updatedGC/August/BipNewman-GC-1mil-GS10-AD5-C007.csv")
BipNewman_GS10_AD5_C014 = createChart("../1milData/giantComponent/updatedGC/August/BipNewman-GC-1mil-GS10-AD5-C014.csv")
BipNewman_GS10_AD5_C021 = createChart("../1milData/giantComponent/updatedGC/August/BipNewman-GC-1mil-GS10-AD5-C021.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Giant Component Newman, Size 1 million Nodes, Average Degree 5, Group Size 10', fontsize=22)

mp.plot(Newman_GS10_AD5_C007, 'ms--', label='Clustering Coef = 0.07')
mp.plot(Newman_GS10_AD5_C014, 'b*--', label='Clustering Coef = 0.14')
mp.plot(Newman_GS10_AD5_C021, 'go--', label='Clustering Coef = 0.21')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
mp.figure(figsize=(25,12))
mp.title('Giant Component, Bipartite Newman, Size 1 million Nodes, Average Degree 5, Group Size 10', fontsize=22)

mp.plot(BipNewman_GS10_AD5_C007[:], 'ms--', color='#af7ac5' ,label='SIR 0.07')
mp.plot(BipNewman_GS10_AD5_C014[:], 'b*--', color='#3498db' ,label='SIR 0.14')
mp.plot(BipNewman_GS10_AD5_C021[:], 'go--', color='#58d68d' ,label='SIR 0.21')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
mp.figure(figsize=(25,12))
mp.title('Newman vs Bipartite Newman', fontsize=22)

mp.plot(Newman_GS10_AD5_C007, 'ms--', label='Newman 0.07')
mp.plot(Newman_GS10_AD5_C014, 'b*--', label='Newman 0.14')
mp.plot(Newman_GS10_AD5_C021, 'go--', label='Newman 0.21')
mp.plot(BipNewman_GS10_AD5_C007[:], 'ms--', color='#af7ac5' ,label='Bipartite 0.07')
mp.plot(BipNewman_GS10_AD5_C014[:], 'b*--', color='#3498db' ,label='Bipartite 0.14')
mp.plot(BipNewman_GS10_AD5_C021[:], 'go--', color='#58d68d' ,label='Bipartite 0.21')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
