```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
TriangleFree_GS50_AD10_4 = createChart("../1milData/giantComponent/updatedGC/August/TriFree-GC-1mil-GS50-AD10-4.csv")
TriangleFree_GS50_AD10_9 = createChart("../1milData/giantComponent/updatedGC/August/TriFree-GC-1mil-GS50-AD10-9.csv")

TriangleFree_GS20_AD10_4 = createChart("../1milData/giantComponent/updatedGC/August/TriFree-GC-1mil-GS20-AD10-4.csv")
TriangleFree_GS20_AD10_9 = createChart("../1milData/giantComponent/updatedGC/August/TriFree-GC-1mil-GS20-AD10-9.csv")
```

```python
Triangle_GS50_AD10_4 = createChart("../1milData/giantComponent/updatedGC/August/Triangle-GC-1mil-GS50-AD10-4.csv")
Triangle_GS50_AD10_9 = createChart("../1milData/giantComponent/updatedGC/August/Triangle-GC-1mil-GS50-AD10-9.csv")

Triangle_GS20_AD10_4 = createChart("../1milData/giantComponent/updatedGC/August/Triangle-GC-1mil-GS20-AD10-4.csv")
Triangle_GS20_AD10_9 = createChart("../1milData/giantComponent/updatedGC/August/Triangle-GC-1mil-GS20-AD10-9.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free, Average Degree 10, In-Deg 4, Giant Component', fontsize=22)

mp.plot(TriangleFree_GS50_AD10_4, 'mo--', label='Group Size 50')
mp.plot(TriangleFree_GS20_AD10_4, 'mo--', color='#af7ac5',label='Group Size 20')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
