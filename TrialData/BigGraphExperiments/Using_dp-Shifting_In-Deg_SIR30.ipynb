```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;

def createChart2(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
TriangleFree_GS50_AD26_10 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-10.csv")
TriangleFree_GS50_AD26_15 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-15.csv")
TriangleFree_GS50_AD26_20 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-20.csv")
TriangleFree_GS50_AD26_25 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-25.csv")

Triangle_GS50_AD26_10 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-10.csv")
Triangle_GS50_AD26_15 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-15.csv")
Triangle_GS50_AD26_20 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-20.csv")
Triangle_GS50_AD26_25 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-25.csv")
```

<h2>Using Double data type as probability p for input in SIR. 30 trials with at least <b>1000</b> people being infected</h2>

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free, SIR30, Average Degree 26', fontsize=22)

mp.plot(TriangleFree_GS50_AD26_10[:.3], 'mo--', color='#ba4a00', label='In-Deg 10')
mp.plot(TriangleFree_GS50_AD26_15[:.3], 'mo--', color='#154360', label='In-Deg 15')
mp.plot(TriangleFree_GS50_AD26_20[:.3], 'bo--', color='#2980b9', label='In-Deg 20')
mp.plot(TriangleFree_GS50_AD26_25[:.3], 'go--', color='#a9cce3', label='In-Deg 25')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
