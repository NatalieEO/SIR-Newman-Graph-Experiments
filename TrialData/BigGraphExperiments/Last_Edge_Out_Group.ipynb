```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;

def createChart2(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
TriFree_GS50_AD26_25_LE50 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-25-LE50.csv")
Tri_GS50_AD26_25_LE50 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-25-LE50.csv")

TriFree_GS50_AD26_25_LE25 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-25-LE25.csv")
Tri_GS50_AD26_25_LE25 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-25-LE25.csv")

TriangleFree_GS50_AD26_25 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-25.csv")
Triangle_GS50_AD26_25 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-25.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free, SIR30, Average Degree 26, In-Degree 25', fontsize=22)

mp.plot(TriFree_GS50_AD26_25_LE25[:.3], 'bo--', color='#a9cce3', label='25%')
mp.plot(TriFree_GS50_AD26_25_LE50[:.3], 'bo--', color='#2980b9', label='50%')
mp.plot(TriangleFree_GS50_AD26_25[:.3], 'go--', color='#154360', label='100%')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
