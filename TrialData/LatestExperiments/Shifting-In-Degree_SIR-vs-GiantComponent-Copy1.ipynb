```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;

def createChart2(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
TriangleFree_SIR_GS50_AD26_15 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-15.csv")
TriangleFree_SIR_GS50_AD26_20 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-20.csv")
TriangleFree_SIR_GS50_AD26_25 = createChart2("../1milData/1mil1Seed1Day/doubleP/TriFree-1SIR-dp-30-1mil-GS50-AD26-25.csv")

Triangle_SIR_GS50_AD26_15 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-15.csv")
Triangle_SIR_GS50_AD26_20 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-20.csv")
Triangle_SIR_GS50_AD26_25 = createChart2("../1milData/1mil1Seed1Day/doubleP/Triangle-1SIR-dp-30-1mil-GS50-AD26-25.csv")
```
