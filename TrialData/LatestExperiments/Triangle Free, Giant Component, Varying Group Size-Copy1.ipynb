```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;

def createChart2(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;

def groupSizeChart(csvFile, startIndex, endIndex, groupSize):
    chart = pd.read_csv(csvFile, sep=',')
    chart = chart[startIndex:endIndex]
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart = chart[['GroupSize', 'PortionInf', 'InfRate']]
    chart = chart.groupby('InfRate').mean()
    chart['GroupSize'] = groupSize
    return chart;

def groupSizeChart2(csvFile, startIndex, endIndex):
    chart = pd.read_csv(csvFile, sep=',')
    chart = chart[startIndex:endIndex]
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart = chart[['GroupSize', 'PortionInf', 'InfRate']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
#Line Plot
gs20 = groupSizeChart2("../1milData/giantComponent/updatedGC/August/TriFree-GC-1mil-GS20-AD10-9.csv", 150, 160)
gs50 = groupSizeChart2("../1milData/giantComponent/updatedGC/August/TriFree-GC-1mil-GS50-AD10-9.csv", 150, 160)
gs100 = groupSizeChart2("../1milData/giantComponent/GC-1mil-TriangleFree-GS100-AD10-9.csv", 150, 160)
gs200 = groupSizeChart2("../1milData/giantComponent/GC-1mil-TriangleFree-GS200-AD10-9.csv", 150, 160)

gs500 = groupSizeChart2("../1milData/giantComponent/GC-Partial-1mil-TriangleFree-GS500-AD10-9.csv", 50, 60)
gs1000 = groupSizeChart2("../1milData/giantComponent/GC-Partial-1mil-TriangleFree-GS1000-AD10-9.csv",50, 60)
gs10000 = groupSizeChart2("../1milData/giantComponent/GC-Partial-1mil-TriangleFree-GS10000-AD10-9.csv",50, 60)

finalDf = pd.concat([gs20, gs50, gs100, gs200, gs500, gs1000, gs10000])
mp.figure(figsize=(25, 12))
mp.title('Population Portion Infected By Group Size', fontsize=22)
mp.plot(finalDf['GroupSize'], finalDf['PortionInf'], 'ro--')
mp.show()
```
