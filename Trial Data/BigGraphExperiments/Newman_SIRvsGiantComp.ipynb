```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
Newman_GS10_AD5_C02 = createChart("../1milData/giantComponent/Summer/Newman-1mil-GS10-AD5-C02.csv")
Newman_GS10_AD5_C04 = createChart("../1milData/giantComponent/Summer/Newman-1mil-GS10-AD5-C04.csv")
Newman_GS10_AD5_C06 = createChart("../1milData/giantComponent/Summer/Newman-1mil-GS10-AD5-C06.csv")
```

```python
Newman_SIR_GS10_AD5_C02 = createChart("../1milData/1mil1SeedsMultiDayData/Summer/Newman-SIR30-1mil-GS10-AD5-C02.csv")
Newman_SIR_GS10_AD5_C04 = createChart("../1milData/1mil1SeedsMultiDayData/Summer/Newman-SIR30-1mil-GS10-AD5-C04.csv")
Newman_SIR_GS10_AD5_C06 = createChart("../1milData/1mil1SeedsMultiDayData/Summer/Newman-SIR30-1mil-GS10-AD5-C06.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Giant Component Newman, Size 1 million Nodes, Average Degree 5, Group Size 10', fontsize=22)

mp.plot(Newman_GS10_AD5_C02, 'ms--', label='Clustering Coef = 0.2')
mp.plot(Newman_GS10_AD5_C04, 'b*--', label='Clustering Coef = 0.4')
mp.plot(Newman_GS10_AD5_C06, 'go--', label='Clustering Coef = 0.6')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
mp.figure(figsize=(25,12))
mp.title('SIR Newman, Size 1 million Nodes, Average Degree 5, Group Size 10', fontsize=22)

mp.plot(Newman_SIR_GS10_AD5_C02[:.20], 'ms--', label='Clustering Coef = 0.2')
mp.plot(Newman_SIR_GS10_AD5_C04[:.20], 'b*--', label='Clustering Coef = 0.4')
mp.plot(Newman_SIR_GS10_AD5_C06[:.20], 'go--', label='Clustering Coef = 0.6')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
