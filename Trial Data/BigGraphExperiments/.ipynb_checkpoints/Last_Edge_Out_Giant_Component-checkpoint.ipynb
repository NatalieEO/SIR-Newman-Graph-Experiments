```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
TriangleFree_GS50_AD26_25_LE50 = createChart("../1milData/giantComponent/July/TriFree-GC-1mil-GS50-AD26-25-LE50.csv")
TriangleFree_GS50_AD26_25_LE25 = createChart("../1milData/giantComponent/July/TriFree-GC-1mil-GS50-AD26-25-LE25.csv")
TriangleFree_GS50_AD26_25 = createChart("../1milData/giantComponent/July/TriFree-GC-1mil-GS50-AD26-25.csv")

Triangle_GS50_AD26_25_LE50 = createChart("../1milData/giantComponent/July/Triangle-GC-1mil-GS50-AD26-25-LE50.csv")
Triangle_GS50_AD26_25_LE25 = createChart("../1milData/giantComponent/July/Triangle-GC-1mil-GS50-AD26-25-LE25.csv")
Triangle_GS50_AD26_25 = createChart("../1milData/giantComponent/July/Triangle-GC-1mil-GS50-AD26-25.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free, SIR30, Average Degree 26, In-Degree 25', fontsize=22)

mp.plot(TriangleFree_GS50_AD26_25_LE25[:], 'bo--', color='#a9cce3', label='25%')
mp.plot(TriangleFree_GS50_AD26_25_LE50[:], 'bo--', color='#2980b9', label='50%')
mp.plot(TriangleFree_GS50_AD26_25[:], 'go--', color='#154360', label='100%')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free, SIR30, Average Degree 26, In-Degree 25', fontsize=22)

mp.plot(Triangle_GS50_AD26_25_LE25[:], 'bo--', color='#edbb99', label='25%')
mp.plot(Triangle_GS50_AD26_25_LE50[:], 'bo--', color='#d35400', label='50%')
mp.plot(Triangle_GS50_AD26_25[:], 'go--', color='#6e2c00', label='100%')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle vs Triangle Free, Average Degree 26, 100%', fontsize=22)

mp.plot(Triangle_GS50_AD26_25[:.7], 'm*--', color='#6e2c00', label='Triangle')
mp.plot(TriangleFree_GS50_AD26_25[:.7], 'mo--', color='#154360', label='Triangle-Free')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')
    
for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
