```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF
```

```python
once = pd.read_csv('../1milData/randTesting/triFree20-1call.csv', sep=',')
loop = pd.read_csv('../1milData/randTesting/triFree20-loopTimeCalls.csv', sep=',')
loopij = pd.read_csv('../1milData/randTesting/triFree20-loopijCalls.csv', sep=',')
triFree1Mil_20 = pd.read_csv('../1milData/triFree20.csv', sep=',')
```

```python
once['PortionInf'] = once['NumInfected']/once['NumNodes']
once['InfRate'] =once['InfRate']/10000
once.tail()
```

```python
loop['PortionInf'] = loop['NumInfected']/loop['NumNodes']
loop['InfRate'] =loop['InfRate']/10000
loop.tail()
```

```python
loopij['PortionInf'] = loopij['NumInfected']/loop['NumNodes']
loopij['InfRate'] =loopij['InfRate']/10000
loopij.tail()
```

```python
triFree1Mil_20['PortionInf'] = triFree1Mil_20['NumInfected']/triFree1Mil_20['NumNodes']
triFree1Mil_20['InfRate'] =triFree1Mil_20['InfRate']/10000
triFree1Mil_20.tail()
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free, Degree=20', fontsize=22)

mp.plot(triFree1Mil_20, 'ko--', label='original')
mp.plot(once, 's--', label='srand seeded once', color='#a8a8a8')
mp.plot(loop, 's--', label='srand seeded in loop with time', color='#a8a8a8')
mp.plot(loopij, 's--', label='srand seeded once', color='#a8a8a8')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
