```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *

import plotly.figure_factory as FF
```

```python
triFree = pd.read_csv('../1milData/giantComponent/triFree-avgd6-gs10-GC-trial1.csv', sep=',')
triangles = pd.read_csv('../1milData/giantComponent/triGraph-avgd6-gs10-GC-trial1.csv', sep=',')
noGroups = pd.read_csv('../1milData/giantComponent/noGroups-avgd6-gs0-GC-trial1.csv', sep=',')
```

```python
triFree['PortionInf'] = triFree['NumInfected']/triFree['NumNodes']
triFree['InfRate'] = triFree['InfRate']/10000

triangles['PortionInf'] = triangles['NumInfected']/triangles['NumNodes']
triangles['InfRate'] = triangles['InfRate']/10000

noGroups['PortionInf'] = noGroups['NumInfected']/noGroups['NumNodes']
noGroups['InfRate'] = noGroups['InfRate']/10000
```

```python
triFree = triFree[['InfRate', 'NumInfected']]
triangles = triangles[['InfRate', 'NumInfected']]
noGroups = noGroups[['InfRate', 'NumInfected']]
```

```python
triFree = triFree.groupby('InfRate').mean()
triangles = triangles.groupby('InfRate').mean()
noGroups= noGroups.groupby('InfRate').mean()
```

```python
mp.figure(figsize=(25,12))
mp.title('Size 1 million Nodes, Average Degree 6', fontsize=22)

mp.plot(triFree, 'ms--', label='Triangle Free')

mp.plot(triangles, 'b*--', label='Triangles')

mp.plot(noGroups, 'go--', label='no Groups')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
