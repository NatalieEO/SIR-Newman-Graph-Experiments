```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
BipNewman_GS10_AD5_C007 = createChart("../1milData/giantComponent/Summer/BipNewman-1mil-GS10-AD5-C007.csv")
BipNewman_GS10_AD5_C014 = createChart("../1milData/giantComponent/Summer/BipNewman-1mil-GS10-AD5-C014.csv")
BipNewman_GS10_AD5_C021 = createChart("../1milData/giantComponent/Summer/BipNewman-1mil-GS10-AD5-C021.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle-Free Newman, Size 1 million Nodes, Average Degree 5, Group Size 10', fontsize=22)

mp.plot(BipNewman_GS10_AD5_C007, 'ms--', label='Clustering Coef = 0.07')
mp.plot(BipNewman_GS10_AD5_C014, 'b*--', label='Clustering Coef = 0.14')
mp.plot(BipNewman_GS10_AD5_C021, 'go--', label='Clustering Coef = 0.21')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
Newman_GS10_AD5_C007 = createChart("../1milData/giantComponent/Summer/Newman-1mil-GS10-AD5-C007.csv")
Newman_GS10_AD5_C014 = createChart("../1milData/giantComponent/Summer/Newman-1mil-GS10-AD5-C014.csv")
Newman_GS10_AD5_C021 = createChart("../1milData/giantComponent/Summer/Newman-1mil-GS10-AD5-C021.csv")
```
