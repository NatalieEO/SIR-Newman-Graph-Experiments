```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF

def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'PortionInf']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

```python
TriangleFree_GS50_AD26_15 = createChart("../1milData/1mil1SeedsMultiDayData/July/TriFree-SIR30-1mil-GS50-AD26-15.csv")
TriangleFree_GS50_AD26_20 = createChart("../1milData/1mil1SeedsMultiDayData/July/TriFree-SIR30-1mil-GS50-AD26-20.csv")
TriangleFree_GS50_AD26_25 = createChart("../1milData/1mil1SeedsMultiDayData/July/TriFree-SIR30-1mil-GS50-AD26-25.csv")

Triangle_GS50_AD26_15 = createChart("../1milData/1mil1SeedsMultiDayData/July/Triangle-SIR30-1mil-GS50-AD26-15.csv")
Triangle_GS50_AD26_20 = createChart("../1milData/1mil1SeedsMultiDayData/July/Triangle-SIR30-1mil-GS50-AD26-20.csv")
Triangle_GS50_AD26_25 = createChart("../1milData/1mil1SeedsMultiDayData/July/Triangle-SIR30-1mil-GS50-AD26-25.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free, SIR30, Average Degree 26', fontsize=22)

mp.plot(TriangleFree_GS50_AD26_15[:.2], 'mo--', color='#154360', label='In-Deg 15')
mp.plot(TriangleFree_GS50_AD26_20[:.2], 'bo--', color='#2980b9', label='In-Deg 20')
mp.plot(TriangleFree_GS50_AD26_25[:.2], 'go--', color='#a9cce3', label='In-Deg 25')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Graph, SIR30, Average Degree 26', fontsize=22)

mp.plot(Triangle_GS50_AD26_15[:.2], 'm*--', color='#6e2c00', label='In-Deg 15')
mp.plot(Triangle_GS50_AD26_20[:.2], 'b*--', color='#d35400', label='In-Deg 20')
mp.plot(Triangle_GS50_AD26_25[:.2], 'g*--', color='#edbb99', label='In-Deg 25')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')
    
for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
