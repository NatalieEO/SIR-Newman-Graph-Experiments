```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF
```

```python
triFree1Mil_20 = pd.read_csv('../1milData/triFree20.csv', sep=',')
triFree1Mil2_20 = pd.read_csv('../1milData/triFree20redo.csv', sep=',')
```

```python
triFree1Mil_20['PortionInf'] = triFree1Mil_20['NumInfected']/triFree1Mil_20['NumNodes']
triFree1Mil_20['InfRate'] =triFree1Mil_20['InfRate']/10000
triFree1Mil_20.tail()
```

```python
triFree1Mil2_20['PortionInf'] = triFree1Mil2_20['NumInfected']/triFree1Mil_20['NumNodes']
triFree1Mil2_20['InfRate'] =triFree1Mil2_20['InfRate']/10000
triFree1Mil2_20.tail()
```

```python
triFree1Mil_20.head()
```

```python
triFree1Mil_20= triFree1Mil_20[['InfRate', 'PortionInf']]
triFree1Mil_20 = triFree1Mil_20.groupby('InfRate').mean()
triFree1Mil_20.head()
```

```python
triFree1Mil2_20= triFree1Mil2_20[['InfRate', 'PortionInf']]
triFree1Mil2_20 = triFree1Mil2_20.groupby('InfRate').mean()
triFree1Mil2_20.head()
```

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free, Degree=20', fontsize=22)

mp.plot(triFree1Mil_20[.05:.25], 'ko--', label='Triangle Free 1 Million')
mp.plot(triFree1Mil2_20[0.05:.25], 's--', label='Triangle Free 1 Million Retrial', color='#a8a8a8')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
