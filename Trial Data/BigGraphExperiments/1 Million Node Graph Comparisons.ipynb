```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *
import plotly.figure_factory as FF
```

```python
oneMilTriFree20 = pd.read_csv('../1milData/updated/triFree20.csv', sep=',')
oneMilTriGraph20 = pd.read_csv('../1milData/updated/triGraph20.csv', sep=',')
oneMilNoGroups20 = pd.read_csv('../1milData/updated/noGroups20.csv', sep=',')


oneMilTriFree40 = pd.read_csv('../1milData/updated/triFree40.csv', sep=',')
oneMilTriGraph40 = pd.read_csv('../1milData/updated/triGraph40.csv', sep=',')
oneMilNoGroups40 = pd.read_csv('../1milData/updated/noGroups40.csv', sep=',')
```

```python
oneMilNoGroups20.tail()
```

```python
oneMilTriFree20['PortionInf'] = oneMilTriFree20['NumInfected']/oneMilTriFree20['NumNodes']
oneMilTriFree20['InfRate'] = oneMilTriFree20['InfRate']/10000

oneMilTriGraph20['PortionInf'] = oneMilTriGraph20['NumInfected']/oneMilTriGraph20['NumNodes']
oneMilTriGraph20['InfRate'] = oneMilTriGraph20['InfRate']/10000

oneMilNoGroups20['PortionInf'] = oneMilNoGroups20['NumInfected']/oneMilNoGroups20['NumNodes']
oneMilNoGroups20['InfRate'] = oneMilNoGroups20['InfRate']/10000




oneMilTriFree40['PortionInf'] = oneMilTriFree40['NumInfected']/oneMilTriFree40['NumNodes']
oneMilTriFree40['InfRate'] = oneMilTriFree40['InfRate']/10000

oneMilTriGraph40['PortionInf'] = oneMilTriGraph40['NumInfected']/oneMilTriGraph40['NumNodes']
oneMilTriGraph40['InfRate'] = oneMilTriGraph40['InfRate']/10000

oneMilNoGroups40['PortionInf'] = oneMilNoGroups40['NumInfected']/oneMilNoGroups40['NumNodes']
oneMilNoGroups40['InfRate'] = oneMilNoGroups40['InfRate']/10000
```

```python
oneMilNoGroups40.tail()
```

```python
oneMilTriFree20= oneMilTriFree20[['InfRate', 'PortionInf']]
oneMilTriFree20 = oneMilTriFree20.groupby('InfRate').mean()

oneMilTriGraph20= oneMilTriGraph20[['InfRate', 'PortionInf']]
oneMilTriGraph20 = oneMilTriGraph20.groupby('InfRate').mean()

oneMilNoGroups20= oneMilNoGroups20[['InfRate', 'PortionInf']]
oneMilNoGroups20 = oneMilNoGroups20.groupby('InfRate').mean()



oneMilTriFree40= oneMilTriFree40[['InfRate', 'PortionInf']]
oneMilTriFree40 = oneMilTriFree40.groupby('InfRate').mean()

oneMilTriGraph40= oneMilTriGraph40[['InfRate', 'PortionInf']]
oneMilTriGraph40 = oneMilTriGraph40.groupby('InfRate').mean()


oneMilNoGroups40= oneMilNoGroups40[['InfRate', 'PortionInf']]
oneMilNoGroups40 = oneMilNoGroups40.groupby('InfRate').mean()
```

```python
triFree1Mil_20 = pd.read_csv('../1milData/triFree20.csv', sep=',')
triFree1Mil_20['PortionInf'] = triFree1Mil_20['NumInfected']/triFree1Mil_20['NumNodes']
triFree1Mil_20['InfRate'] =triFree1Mil_20['InfRate']/10000
triFree1Mil_20= triFree1Mil_20[['InfRate', 'PortionInf']]
triFree1Mil_20 = triFree1Mil_20.groupby('InfRate').mean()
triFree1Mil_20.head()
```

<h1>1 Million Nodes, Average Degree 20</h1>

```python
mp.figure(figsize=(25,12))
mp.title('Triangle Free vs Triangles vs No Groups, avgDeg=20, 1 Million Nodes', fontsize=22)
mp.plot(oneMilTriFree20[:.3], 's--', label='Triangle Free', color ='#2e86c1')
mp.plot(oneMilTriGraph20, 'ko--', label='Triangles', color='#009d00')
mp.plot(oneMilNoGroups20, 'm*--', label='No Groups', color='#ff4e4e')

#mp.plot(triFree1Mil_20[:.3], 'ko--', label='triangle free original')
legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
