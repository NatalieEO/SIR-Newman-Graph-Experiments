```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *

import plotly.figure_factory as FF
```

```python
def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'NumInfected']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

<h2>Try to guarantee 30 valid trials, but go to next infection rate after 1000 attempts.</h2>

```python
noGroups5 = createChart("../1milData/1mil1SeedsMultiDayData/Summer/NoGroups-SIR30-1mil-GS0-AD5.csv")
triangle5 = createChart("../1milData/1mil1SeedsMultiDayData/Summer/Triangle-SIR30-1mil-GS8-AD5-4.csv")
triFree5 = createChart("../1milData/1mil1SeedsMultiDayData/Summer/TriFree-SIR30-1mil-GS8-AD5.csv")
```

```python
mp.figure(figsize=(25,12))
mp.title('Size 1 million Nodes, Average Degree 15, Group Size 30', fontsize=22)

mp.plot(triFree5[:], 'ms--', label='Triangle Free')

mp.plot(triangle5[:], 'b*--', label='Triangles')

mp.plot(noGroups5[:], 'go--', label='no Groups')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
