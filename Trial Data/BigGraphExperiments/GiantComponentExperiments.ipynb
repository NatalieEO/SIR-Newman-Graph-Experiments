```python
import pandas as pd
import plotly
import matplotlib.pyplot as mp
import plotly.plotly as py 
import plotly.graph_objs as go
import numpy as np
from plotly.graph_objs import *

import plotly.figure_factory as FF
```

```python
def createChart(csvFile):   
    chart = pd.read_csv(csvFile, sep=',')
    chart['PortionInf'] = chart['NumInfected']/chart['NumNodes']
    chart['InfRate'] = chart['InfRate']/10000
    chart = chart[['InfRate', 'NumInfected']]
    chart = chart.groupby('InfRate').mean()
    return chart;
```

<h1>Average Degree 5, Group Size 8, 1 Million Nodes</h1>

```python
noGroups5 = createChart("../1milData/giantComponent/Summer/NoGroups-1mil-GS0-AD5.csv")
triangle5 = createChart("../1milData/giantComponent/Summer/Triangle-1mil-GS8-AD5-4.csv")
triFree5 = createChart("../1milData/giantComponent/Summer/TriFree-1mil-GS8-AD5.csv")
```

```python
noGroups5.head()
```

```python
mp.figure(figsize=(25,12))
mp.title('Size 1 million Nodes, Average Degree 5, Group Size 8', fontsize=22)

mp.plot(triFree5[0.4:], 'ms--', label='Triangle Free')

mp.plot(triangle5[0.4:], 'b*--', label='Triangles')

mp.plot(noGroups5[0.4:], 'go--', label='no Groups')

legend = mp.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)
mp.show()
```
